```python
import pandas as pd
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv")
df
```

```python
duplicate_counts = df.duplicated().sum()

print(duplicate_counts)
df_no_duplicates = df.drop_duplicates()

df_no_duplicates
```

```python
duplicate_counts = df['Respondent'].duplicated().sum()

print(duplicate_counts)
df_no_duplicates = df.drop_duplicates()

df_no_duplicates
```

```python
blank_rows_count = df['EdLevel'].isna().sum()
blank_rows_count
```

```python
majority_category = df['Employment'].mode()

print("Majority category in 'Employment' column:")
print(majority_category)
```

```python
undergrad_major_counts = df['UndergradMajor'].value_counts()

# Find the category with the minimum number of rows
category_with_minimum_rows = undergrad_major_counts.idxmin()
category_with_minimum_rows
```

```python
compfreq_value_counts = df['CompFreq'].value_counts()

total_values_in_compfreq = compfreq_value_counts.sum()
total_values_in_compfreq
```

```python
df_no_duplicates = df.drop_duplicates()

# Filter the DataFrame to include only respondents who are paid yearly
respondents_paid_yearly = df_no_duplicates[df_no_duplicates['CompFreq'] == 'Yearly']

# Count the number of respondents being paid yearly
number_of_respondents_paid_yearly = respondents_paid_yearly.shape[0]

print("Number of respondents being paid yearly after removing duplicates:", number_of_respondents_paid_yearly)
```
